```python
import pandas as pd
pd.set_option("max_columns", 50)
```

```python
diab_data = pd.read_csv("c:data/diabetic_data.csv")
```

```python
dispo_desc = pd.read_csv("c:data/IDs_mapping.csv")
```

```python
# capture index of where discharge disposition codes start
disch_type_start = dispo_desc[dispo_desc["admission_type_id"] == "discharge_disposition_id"].index.values.astype(int)[0]

# capture index of where admission disposition codes start
admit_src_start = dispo_desc[dispo_desc["admission_type_id"] == "admission_source_id"].index.values.astype(int)[0]
```

```python
admit_type = dispo_desc.iloc[0:disch_type_start-1] # subtract 1 because there is a blank line before each disposition group

disch_disp = dispo_desc.iloc[disch_type_start+1:admit_src_start-1]

admit_src = dispo_desc.iloc[admit_src_start+1:]
```

```python
admit_src.head()
```

```python
admit_type_dict = admit_type.set_index("admission_type_id").to_dict()["description"]
disch_disp_dict = disch_disp.set_index("admission_type_id").to_dict()["description"]
admit_src_dict = admit_src.set_index("admission_type_id").to_dict()["description"]

admit_src_dict = {int(k):v for k,v in admit_src_dict.items()}
admit_type_dict = {int(k):v for k,v in admit_type_dict.items()}
disch_disp_dict = {int(k):v for k,v in disch_disp_dict.items()}
```

```python
diab_data['admission_source'] = diab_data.admission_source_id.map(admit_src_dict)
diab_data['discharge_disposition'] = diab_data.discharge_disposition_id.map(disch_disp_dict)
diab_data['admission_type'] = diab_data.admission_type_id.map(admit_src_dict)
```

```python
diab_data.head()
```

```python
del diab_data['admission_source_id']
del diab_data['admission_type_id']
del diab_data['discharge_disposition_id']
```

```python
ques_list = []
for col in list(diab_data.columns):
    if '?' in set(diab_data[col]):
        #print(col)
        ques_list.append(col)
```

```python
ques_list
```

```python
for col in ques_list:
    diab_data[col] = diab_data[col].replace('?', '')
```

```python
for col in list(diab_data.columns):
    if '?' in set(diab_data[col]):
        print(col)
```

```python
# write data to csv file
diab_final = diab_data.to_csv("c:data/no_code_data.csv")
```

```python
diab_final = pd.read_csv("data/no_code_data.csv")
```

```python
diab_final['readmitted'] = diab_final['readmitted'].map({">30": "YES", "<30":"YES", "NO":"NO"})
```

```python
diab_final.head()
```

```python
# columns need to be deleted since they are not adding any value
del_cols = ["Unnamed: 0", "encounter_id", "patient_nbr", "weight", "discharge_disposition", 
            "admission_source", "payer_code", "medical_specialty", "admisison_type"]

for col in del_cols:
    if col in diab_final.columns:
        diab_final.drop(col, inplace=True, axis = 1)
        #del diab_final[col]
```

```python
set(diab_final.readmitted)
```

```python
# capture all the columns with character data type
col_dtype = diab_final.columns.to_series().groupby(diab_final.dtypes).groups
char_cols = {k.name: v for k, v in col_dtype.items()}
char_cols = list(char_cols['object']) # list of character columns
# strip leading and trailing spaces
for col in char_cols:
    diab_final[col] = diab_final[col].str.strip()
```
